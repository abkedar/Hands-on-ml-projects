```python
## Without a constant
import statsmodels.api as sm

X = df["RM"]
y = target["MEDV"]

# Note the difference in arguments order
model = sm.OLS(y, X).fit()
predictions = model.predict(X) # make the predictions by the model

# Print out the statistics
model.summary()
```

```python
X = df["RM"]    # X usually means our input variables (or independent variables)
y = target["MEDV"] ## Y usually means our output/dependent variable
X = sm.add_constant(X) # let's add an intercept (bets_0) to our model

# Note the difference in arguments order
model = sm.OLS(y, X).fit() ## sm.ols(output, input)
predictions = model.predict(X)

# Predict out the statistics
model.summary()
```

```python
X = df[["RM", "LSTAT"]]
y = target["MEDV"]

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

model.summary()
```

## Linear Regression in skLearn

```python
from sklearn import linear_model
```

```python
from sklearn import datasets ## import datasets from scikit-learn
data = datasets.load_boston() ## loads Boston datasets from datasets libracy
```

```python
# define the data/predictors as the pre-set feature names
df = pd.DataFrame(data.data, columns=data.feature_names)

# Put the trget (housing value -- MEDV) in another DataFrame
target = pd.DataFrame(data.target, columns = ["MEDV"])
```

```python
X = df
y = target["MEDV"]

lm = linear_model.LinearRegression()
model = lm.fit(X, y)
```

```python
model
```

```python
predictions = lm.predict(X)
print(predictions)[0:5]
```

```python
lm.score(X, y)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
X = df[["RM", "LSTAT"]]
y = target["MEDV"]
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.ylabel("dependent variable")
plt.xlabel("independent variable")
plt.show()
```
